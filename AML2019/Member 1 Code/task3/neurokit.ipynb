```python
def print_structure(key, x, prefix=''):
    print("%s- %s: %s" % (prefix, key, type(x)))    
    if isinstance(x, dict):
        for key in x:
            print_structure(key, x[key], prefix+' ')
```

```python
idx = []
X = []
y = []
bad = []
for i in range(len(signals)):
    try:
        s = signals[i]
        ecg = neurokit.ecg_process(s, rsp=None, sampling_rate=300, filter_type='FIR', filter_band='bandpass', filter_frequency=[3, 45], quality_model='default', hrv_features=['time', 'frequency'], age=None, sex=None, position=None)
        vec = [ecg['ECG']['Average_Signal_Quality']]
        for f in hrvfeatures:
            vec.append(hrv[f])
        X.append(vec)
        y.append(labels[i])
        idx.append(i)
        print("============ APPENDED %d ===========" % i)
    except:
        bad.append(i)
print(X[:10])
print(y[:10])
print(idx[:10])
```

```python
hrvs = []
for idx, s in enumerate(signals):
    rpeaks = ecg.christov_segmenter(s, sampling_rate=300.0)[0]
    hrv = neurokit.ecg_hrv(rpeaks, sampling_rate=300.0)
    hrvs.append(hrv)
    if idx % 100 == 0:
        print("processed %d" % idx)
```

```python
import numpy
features = []
for key in hrvs[0]:
    if type(hrvs[0][key]) in [float, numpy.float64, numpy.float32]:
        features.append(key)
features
```

```python
with open("data/hrvfeatures.csv", "w") as f:
    s = "Id"
    for feat in features:
        s += ",%s" % feat
    s += "\n"
    f.write(s)
    for idx, hrv in enumerate(hrvs):
        s = str(idx)
        for feat in features:
            s += ",%f" % hrv.get(feat, np.nan)
        s += '\n'
        f.write(s)
```

```python
X = []
for hrv in hrvs:
    x = []
    for f in features:
        x.append(hrv.get(f, np.nan))
    X.append(x)
```

```python
y = []
with open("data/y_train.csv") as f:
    for line in f.readlines()[1:]:
        y.append(int(line.split(',')[1]))
```

```python
import xgboost as xgb
import numpy as np
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
train_X, eval_X, train_Y, eval_Y = train_test_split(X, y, test_size=0.3)
xg_train = xgb.DMatrix(train_X, label=train_Y)
xg_eval = xgb.DMatrix(eval_X, label=eval_Y)

# setup parameters for xgboost
param = {}
# use softmax multi-class classification
param['objective'] = 'multi:softmax'
# scale weight of positive examples
param['eta'] = 0.2
param['gamma'] = 1.0
param['max_depth'] = 6
param['silent'] = 1
param['subsample'] = 0.8
param['colsample_bytree'] = 0.9
param['min_child_weight'] = 20
param['num_class'] = 4


watchlist = [(xg_train, 'train'), (xg_eval, 'eval')]
num_round = 60
bst = xgb.train(param, 
                xg_train, 
                num_round, 
                watchlist, 
                feval=lambda y,t: ("f1", f1_score(y, t.get_label(), average='micro')))

# get prediction
pred = bst.predict(xg_eval)
# error_rate = np.sum(pred != test_Y) / test_Y.shape[0]
F1 = f1_score(eval_Y, pred, average='micro')
print('Test error using softmax = {}'.format(F1))
```
