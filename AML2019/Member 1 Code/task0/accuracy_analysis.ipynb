```python
trainset = open("train.csv").readlines()
```

```python
trainset = [x.strip().split(',')[1:] for x in trainset[1:]]
```

```python
trainset
```

```python
prediction = []
for l in trainset:
    floats = [float(x) for x in l[1:]]
    prediction.append(sum(floats)/len(floats))
prediction
```

```python
from statistics import mean
from math import sqrt
sqrt(mean([(prediction[i] - float(trainset[i][0]))**2 for i in range(len(prediction))]))
```

```python
for i in range(len(prediction)):
    if str(prediction[i]) != trainset[i][0]:
        print(i, prediction[i], trainset[i][0], prediction[i] - float(trainset[i][0]))
```
