```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
```

```python
X = pd.read_csv("{}/data/X_train_first_10.csv".format(dir_cur), index_col=0).values
x = X[0]
```

```python
result = ecg.ecg(X[4][~np.isnan(X[4])], SAMPLING_RATE, show=0)
```

```python
param_description = [
    'Signal time axis reference (seconds).',
    'Filtered ECG signal.',
    'R-peak location indices.',
    'Templates time axis reference (seconds).',
    'Extracted heartbeat templates.',
    'Heart rate time axis reference (seconds).',
    'Instantaneous heart rate (bpm).',
]

param_name = [
    'ts',
    'filtered',
    'rpeaks',
    'templates_ts',
    'templates',
    'heart_rate_ts',
    'heart_rate',
]
```

```python
fig, axs = plt.subplots(len(result), figsize=(6,3*len(param_description)))
fig.suptitle('ECG result')
for ax, param, title, ylabel in zip(
    axs, result, param_description, param_name):
    ax.plot(param)
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    
plt.tight_layout()
plt.savefig('test.png', dpi=200)
```
